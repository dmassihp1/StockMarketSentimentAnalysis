```python
import re
import spacy
import nltk
nltk.download('stopwords')
nltk.download('punkt')
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords

# Remove stopwords, url's, usernames, hashtag symbols, and convert to lowercase
sw = stopwords.words('english')

def preProcessTweet(tweet):
    tweet = tweet.lower() # convert text to lower-case
    tweet = re.sub('((www\.[^\s]+)|(https?://[^\s]+))', 'URL', tweet) # remove URLs
    tweet = re.sub('@[^\s]+', 'AT_USER', tweet) # remove usernames
    tweet = re.sub(r'#([^\s]+)', r'\1', tweet) # remove the # in #hashtag
    tweet = word_tokenize(tweet) # remove repeated characters (helloooooooo into hello)
    return ' '.join([word for word in tweet if word not in sw])
```

```python
preProcessTweet(mergedDfDupDropped['Tweet'].loc[0])
```

```python
mergedDfDupDropped['CleanedTweets'] = ""
for i, row in mergedDfDupDropped.iterrows():
  mergedDfDupDropped.at[i, 'CleanedTweets'] = preProcessTweet(row.loc["Tweet"])
```

```python
mergedDfDupDropped = mergedDfDupDropped.sample(2000, random_state=42)
```

```python
# Use python textblob package and pretrained bert sentiment model to assign sentiments to
#  tweets (naive bayes classifier model). Take average of both scores as final sentiment score

mergedDfDupDropped["SentimentTextBlob"] = 0
mergedDfDupDropped["SentimentFinBert"] = 0
mergedDfDupDropped['SentimentAvg'] = 0
for i, row in mergedDfDupDropped.iterrows():
  currTweet = row.loc['CleanedTweets']
  currScoreTextBlob = TextBlob(currTweet).sentiment.polarity
  currScoreFinbert = finBertSentiment(currTweet)[0]['score']

  mergedDfDupDropped.at[i,'SentimentTextBlob'] = currScoreTextBlob
  mergedDfDupDropped.at[i,'SentimentFinbert'] = currScoreFinbert
  mergedDfDupDropped.at[i,'SentimentAvg'] = (currScoreFinbert + currScoreTextBlob) / 2


```

### Models to predict stock prices based on sentiment scores


```python
# Set df index to the date, order by date and shift adjusted closing price
# (the predictor variable for our model) to the
#  previous day (so that we are using tweets/sentiment from previous day to predict stock price)

mergedDfDupDropped = mergedDfDupDropped.sort_values(by=['DateOnly'], ascending=True)
mergedDfDupDropped.set_index('DateOnly', inplace=True)

mergedDfDupDropped['Adj Close Next'] = mergedDfDupDropped.groupby('Stock Name')['Adj Close'].shift(-1)

mergedDfDupDropped.dropna(inplace = True)


mergedDfDupDropped
```

```python
mergedDfDupDropped["Stock Name"].value_counts()
```

```python
# We will only use stocks with the 10 highest counts (otherwise will cause errors
#  when building model since we will split into train, validation and tet set, and need
#  enough examples in each set

mask = mergedDfDupDropped['Stock Name'].isin(mergedDfDupDropped["Stock Name"].value_counts().index[0:10].tolist())
newDf = mergedDfDupDropped[mask]
newDf["Stock Name"].value_counts()
```

```python
# To avoid look-ahead bias, use first 70% of data for the training set and the rest as
# the validation and test set. One hot encode the company names


X = newDf[['SentimentAvg', 'Stock Name', 'Open', 'High', 'Low', 'Volume', 'Adj Close']]
Y = newDf['Adj Close Next']

X = pd.get_dummies(X)

splitInd = int(0.7*len(X))

X_train = X[0:splitInd]
Y_train = Y[0:splitInd]

dataLeft = len(X) - len(X_train)

X_val = X[splitInd:splitInd+int(dataLeft/2)]
Y_val = Y[splitInd:splitInd+int(dataLeft/2)]

X_test = X[splitInd + int(dataLeft/2):]
Y_test = Y[splitInd + int(dataLeft/2):]
```

```python
print(len(X), len(X_train), len(X_val), len(X_test))
```

```python
# Random Forest:
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error

def objective(trial):

    params = {
              "n_estimators" : trial.suggest_int("n_estimators", 20, 1000),
              "max_depth" : trial.suggest_int("max_depth", 5, 100),
              "min_samples_split" : trial.suggest_int("min_samples_split", 2, 32),
              "min_samples_leaf" : trial.suggest_int("min_samples_leaf", 1, 32),
              "max_features" : trial.suggest_categorical("max_features", ['sqrt', 'log2']),
              "criterion" : trial.suggest_categorical('criterion', ["squared_error", "absolute_error"])
              # criterion desecribs function to measure quality of the split

    }

    model = RandomForestRegressor(**params)

    model.fit(X_train, Y_train)
    predictions = model.predict(X_val)

    # Metric  to optimize
    rmse = mean_squared_error(Y_val, predictions, squared=False)
    return rmse
```

```python
study = optuna.create_study(direction='minimize', sampler=optuna.samplers.RandomSampler(seed=42))
study.optimize(objective, n_trials=50)

# Print the best parameters found
print("Best trial:")
trial = study.best_trial
```

```python
# Xgboost:

import xgboost as xgb
from sklearn.metrics import mean_squared_error
import optuna

def objective2(trial):
    params = {
        "objective": "reg:squarederror",
        "n_estimators": trial.suggest_int("n_estimators", 20, 1000),
        "learning_rate": trial.suggest_float("learning_rate", 1e-3, .5, log=True),
        "max_depth": trial.suggest_int("max_depth", 10, 100),
        "subsample": trial.suggest_float("subsample", 0.05, 1.0),
        "colsample_bytree": trial.suggest_float("colsample_bytree", 0.05, 1.0),
        "min_child_weight": trial.suggest_int("min_child_weight", 1, 30),
    }

    model = xgb.XGBRegressor(**params)
    model.fit(X_train, Y_train, verbose=False)
    predictions = model.predict(X_val)
    rmse = mean_squared_error(Y_val, predictions, squared=False)
    return rmse
```

```python
study2 = optuna.create_study(sampler=optuna.samplers.RandomSampler(seed=42))
study2.optimize(objective2, n_trials=50)

# Print the best parameters found
print("Best trial:")
trial = study2.best_trial
```

```python
# Print best hyperparameters and rmse (square root of the average of squared differences between
#  predictions and actual values)

print('Best hyperparameters random forest:', study.best_params)
print('Best RMSE random forest:', study.best_value)
```

```python
print('Best hyperparameters xgboost:', study2.best_params)
print('Best RMSE xgboost:', study2.best_value)
```

```python
# Calculate MAPE (mean absolute percentage error for both models/average of how far off
#  predictions are on average) and RMSE

from sklearn.metrics import mean_absolute_percentage_error

model1 = RandomForestRegressor(**study.best_params)
model1.fit(X_train, Y_train)
model1Preds = model1.predict(X_test)
print("MAPE for random forest: ", mean_absolute_percentage_error(Y_test, model1Preds))
print("RMSE for random forest: ", mean_squared_error(Y_test, model1Preds, squared=False))


model2 = xgb.XGBRegressor(**study2.best_params)
model2.fit(X_train, Y_train)
model2Preds = model2.predict(X_test)
print("MAPE for xgboost: ", mean_absolute_percentage_error(Y_test, model2Preds))
print("RMSE for xgboost: ", mean_squared_error(Y_test, model2Preds, squared=False))
```
